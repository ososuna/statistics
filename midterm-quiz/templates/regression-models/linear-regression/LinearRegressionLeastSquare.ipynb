# Linear regression (y = a + bx)

## Least square method

```python
import numpy as np
from numpy.linalg import inv 
import matplotlib.pyplot as plt
```

### Reading data

```python
x = [16, 32, 48, 56, 64, 80]
y = [10, 15, 20, 22, 30, 32]
```

```python
n = len(x)
```

### Finding required sum for least square methods

```python
sumX,sumX2,sumY,sumXY = 0,0,0,0
for i in range(n):
    sumX = sum(x)
    sumX2 = sum([i ** 2 for i in x])
    sumY = sum(y)
    sumXY = sum([x * y for (x, y) in zip(x, y)])
```

### Obtaining a and b

```python
b = (n * sumXY - sumX * sumY)/(n*sumX2 - sumX * sumX)
a = (sumY - b*sumX)/n
```

### Displaying coefficients a, b & equation

```python
print("\nCoefficients are:")
print("a: ", a)
print("b: ", b)
```

```python
print("\nEquation is:")
print(f'y = {a} + {b}x')
```

### Plot regression model

```python
y_pred = []
for i in range(n):
  y_pred.append(a+b*x[i])
```

```python
plt.scatter(x, y)
plt.plot(x, y_pred, c='blue', linewidth=2)
plt.show()
```
